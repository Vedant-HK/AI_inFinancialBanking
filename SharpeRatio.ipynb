```python
#Calculating sharpe ratios for investment portfolio

import numpy as np
import matplotlib.pyplot as plt

# Annual returns of 3 portfolios
returns_A = [0.12, 0.10, 0.08, 0.11, 0.09]
returns_B = [0.18, 0.15, 0.12, 0.16, 0.14]
returns_C = [0.09, 0.08, 0.07, 0.10, 0.09]

# Convert to numpy arrays
returns_A = np.array(returns_A)
returns_B = np.array(returns_B)
returns_C = np.array(returns_C)

# Annual risk-free rate (5%)
risk_free_rate = 0.05

# Sharpe Ratio function (annual)
def calculate_sharpe(returns, rf):
    avg_return = np.mean(returns)
    std_dev = np.std(returns)
    return (avg_return - rf) / std_dev

# Calculate Sharpe Ratios
sharpe_A = calculate_sharpe(returns_A, risk_free_rate)
sharpe_B = calculate_sharpe(returns_B, risk_free_rate)
sharpe_C = calculate_sharpe(returns_C, risk_free_rate)

print("Annual Sharpe Ratio Portfolio A:", sharpe_A)
print("Annual Sharpe Ratio Portfolio B:", sharpe_B)
print("Annual Sharpe Ratio Portfolio C:", sharpe_C)

# Plot annual returns
plt.figure()
plt.plot(returns_A, marker='o', label='Portfolio A')
plt.plot(returns_B, marker='o', label='Portfolio B')
plt.plot(returns_C, marker='o', label='Portfolio C')

plt.xlabel("Years")
plt.ylabel("Annual Returns")
plt.title("Annual Returns of Three Portfolios")
plt.legend()
plt.show()
```
