```python
import pandas as pd
import numpy as np

data = {
    'Date': pd.to_datetime(['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05']),
    'Price': [100, 105, 110, 115, 120]
}

df = pd.DataFrame(data).set_index('Date')

short_window = 2
long_window = 3

df['SMA_Short'] = df['Price'].rolling(window=short_window).mean()
df['SMA_Long'] = df['Price'].rolling(window=long_window).mean()

df['Signal'] = 0.0
df['Signal'][short_window:] = np.where(df['SMA_Short'][short_window:] > df['SMA_Long'][short_window:], 1.0, 0.0)

df['Positions'] = df['Signal'].diff()

initial_capital = 10000
portfolio = initial_capital
shares = 0

for i in range(len(df)):
    if df['Positions'][i] == 1:
        shares = portfolio //df['Price'][i]
        portfolio -= shares * df['Price'][i]
        print(f"Buy {shares} shares at {df['Price'][i]} on {df.index[i]}")

    elif df['Positions'][i] == -1:
      portfolio += shares * df['Price'][i]
      print(f"sell {shares} shares at {df['Price'][i]} on {df.index[i]}")
      shares = 0

portfolio += shares * df['Price'].iloc[-1]

print(f"\nFinal Portfolio Value:",portfolio)
print("Profit/Loss:", portfolio - initial_capital)

print("\nDataFrame with signals:")
print(df)
```
